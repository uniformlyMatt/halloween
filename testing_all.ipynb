```python
import time
import threading
import numpy as np
```

Functions needed

```python
class Elevator:
    def __init__(self, pin):
        self.pin = pin
        self.ready = threading.Event()
        
    def __str__(self):
        return 'Elevator object on pin {}'.format(self.pin)
        
    def run(self):
        """ Make the elevator go up a certain length, then come back down. """
        
        #gpio.output(motor_in1, gpio.HIGH)
        #gpio.output(motor_in2, gpio.LOW)
        print("going up")
        
        # Wait 1 second before coming back down
        time.sleep(1)
        
        #gpio.output(motor_in1, gpio.LOW)
        #gpio.output(motor_in2, gpio.HIGH)
        print("coming back down")
        
        time.sleep(1)
        
        print('Round trip finished - back to start position')
        self.ready.set()
        
        return 0
    
class Button:
    def __init__(self, pin):
        self.pin = pin
        self.ready = threading.Event()
        
    def __str__(self):
        return 'Button object on pin {}'.format(self.pin)
    
    def run(self):
        """ Wait for the button to be pressed. """
        
        time.sleep(3)
#         while True:
#             if gpio.input(self.pin) == gpio.HIGH:
#                 self.ready.set()
#                 break
        self.ready.set()
        print('\nButton pressed')
        return None
        
def distance():
    """ Distance function for the ultrasonic sensor. """
    
    test_time = 0.002
    
#     # Send out a pulse for 0.00001 s
#     gpio.output(trig, 1)
#     sleep(pulse_delay)
#     gpio.output(trig, 0)
    
#     start_time = time()
#     stop_time = time()
    
#     # get the time of the pulse being sent
#     while gpio.input(echo) == 0:
#         start_time = time()
        
#     # get the time of the pulse returning
#     while gpio.input(echo) == 1:
#         stop_time = time()
        
#     # find the round-trip time of the pulse
#     elapsed_time = stop_time - start_time
    
#     # multiply by the speed of sound (34300 cm/s)
#     # and divide by 2 (there and back)
    
#     return elapsed_time*34300/2
    return test_time*34300/2

class Dispenser:
    def __init__(self, pin, steps = 3600, step_delay = 0.0000001):
        self.pin = pin
        self.steps = steps           # Number of steps needed to rotate the auger
        self.step_delay = step_delay # Extremely short delay between steps
        self.ready = threading.Event()
        
    def __str__(self):
        return 'Dispenser object on pin {}'.format(self.pin)
    
    def run(self):
        """ Dispense candy """
        
#         for i in range(self.steps):
#             gpio.output(ena, gpio.HIGH)
#             #print('Controller enabled')

#             gpio.output(pul, gpio.HIGH)
#             sleep(step_delay)
#             gpio.output(pul, gpio.LOW)
#             sleep(step_delay)

#             gpio.output(ena, gpio.LOW)
        print('Rotating the auger...')
        time.sleep(3)        
        
        self.ready.set()
        print('Candy dispensed')
        
        return None
    
def thread_it(obj, timeout = 10):
    """ General function to handle threading for the physical components of the system. """
    
    thread = threading.Thread(target = obj.run())
    thread.start()
    
    # Run the 'run' function in the obj
    obj.ready.wait(timeout = timeout)
    
    # Clean up
    thread.join()
    obj.ready.clear()
```
